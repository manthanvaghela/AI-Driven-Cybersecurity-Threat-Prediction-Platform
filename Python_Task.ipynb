Write a Python program that takes a sentence and: Counts vowels and consonants and reverses the sentence and converts it to camel case

```python
import re

def analyze_sentence(text):
    """
    Analyzes a sentence to count vowels and consonants, reverse the text,
    and generate a camelCase version.

    Args:
        text (str): Input sentence.

    Returns:
        tuple: (vowel_count, consonant_count, reversed_text, camel_case_text)
    """
    # Define vowel set for quick lookup
    vowel_set = set("aeiouAEIOU")

    # Count vowels and consonants
    vowel_total = sum(1 for ch in text if ch in vowel_set)
    consonant_total = sum(1 for ch in text if ch.isalpha() and ch not in vowel_set)

    # Reverse the string
    reversed_text = text[::-1]

    # Convert to camelCase
    tokens = re.findall(r"[A-Za-z0-9]+", text.lower())
    if tokens:
        camel_case_text = tokens[0] + ''.join(word.capitalize() for word in tokens[1:])
    else:
        camel_case_text = ""

    return vowel_total, consonant_total, reversed_text, camel_case_text


# Example usage
sample_text = "This is a sample sentence."
vowels, consonants, reversed_text, camel_text = analyze_sentence(sample_text)

print(f"Original sentence: {sample_text}")
print(f"Vowel count: {vowels}")
print(f"Consonant count: {consonants}")
print(f"Reversed sentence: {reversed_text}")
print(f"Camel case sentence: {camel_text}")
```

 Create a tuple of 10 random numbers. Then:
Find the max and min
and Count how many times each number appears and Convert it to a list and sort it


```python
import random
from collections import Counter

numbers = tuple(random.randrange(1, 101) for _ in range(10))
print(f"Original tuple: {numbers}")

highest = max(numbers)
lowest = min(numbers)
print(f"Maximum number: {highest}")
print(f"Minimum number: {lowest}")

freq = Counter(numbers)
print(f"Number counts: {freq}")

sorted_list = sorted(numbers)
print(f"Sorted list: {sorted_list}")
```

Generate a 5Ã—5 matrix of random integers using NumPy. Then:
Find the mean, median, and standard deviation and plot using seaborn


```python
import numpy as np

matrix = np.random.randint(low=0, high=101, size=(5, 5))
print(matrix)
```

```python
avg = np.average(matrix)
med = np.median(matrix)
std_dev = np.std(matrix)

print(f"Mean of the matrix: {avg}")
print(f"Median of the matrix: {med}")
print(f"Standard deviation of the matrix: {std_dev}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(6, 5))
sns.heatmap(matrix, ax=ax, cmap="viridis", annot=True, fmt="d")
ax.set_title("Heatmap of Random Matrix")
ax.set_xlabel("Column Index")
ax.set_ylabel("Row Index")
plt.tight_layout()
plt.show()
```

 Prompt:Create a class CyberThreatDetector with:
Attributes: source, threat_level
and Method to classify threat as low/medium/high
and Method to display threat summary
in this way display oops concept example

```python
class ThreatAnalyzer:
    def __init__(self, origin, level_value):
        self.origin = origin
        self.level_value = level_value
        self.level = self._evaluate_level()

    def _evaluate_level(self):
        if self.level_value < 40:
            return "Low"
        if self.level_value < 70:
            return "Medium"
        return "High"

    def show_summary(self):
        print(f"Threat Source: {self.origin}")
        print(f"Threat Level: {self.level}")
```

```python
detector_a = ThreatAnalyzer("Network Scan", 55)
detector_a.show_summary()

print("-" * 20)

detector_b = ThreatAnalyzer("Phishing Email", 85)
detector_b.show_summary()
```
